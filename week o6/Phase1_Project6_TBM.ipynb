**`Background:`**

A Tunnel Boring Machine (TBM) is a complex piece of industrial machinery used to excavate tunnels with a circular cross-section. Dozens of sensors continuously monitor operational parameters. The key performance metric is the **Penetration Rate (PR)** — how fast the cutter head advances through rock per revolution (mm/rev). Maximizing PR while avoiding equipment damage is the central engineering challenge.

**`Problem Statement:`**

Real-world TBM sensor data is inherently **noisy, incomplete, and unlabeled**. Traditional supervised learning fails because engineers rarely annotate each boring cycle with a class label. This project applies **Weak Supervision** — a modern ML technique where heuristic rules (called *labeling functions*) replace manual annotation — to build a classification model that predicts whether PR is **Low**, **Medium**, or **High** based on sensor readings.

**`Questions This Notebook Answers:`**

**`Q01:`** What are the most informative TBM sensor parameters for predicting Penetration Rate?

**`Q02:`** How does weak supervision perform compared to a fully labeled baseline?

**`Q04:`** What operational recommendations can be drawn from the model's feature importance rankings?

**`Q05:`** Where does the model struggle, and what does the confusion matrix reveal about class boundaries?

**`Assumptions:`**
- Synthetic data is generated to closely mimic real TBM sensor distributions from published geomechanical literature
- Missing sensor readings follow a Missing at Random (MAR) pattern — typical of sensor dropout events
- The Q-Value (rock quality index) is treated as a known geological parameter measured at the face

**`Methodology:`**
- Generate a realistic 800-sample synthetic TBM dataset with deliberate noise and missing values
- Conduct thorough EDA to understand feature distributions and inter-variable relationships

<div style="background-color:#E6F2FF; padding:10px; border-radius:8px; box-shadow:0 1px 3px 0 rgba(0,0,0,0.1); border:1px solid #66C2A5; margin-top:10px; margin-left:20px;">

        3.1 Synthetic TBM Sensor Data
    </h2>
</div>

In a real-world deployment, this data would come from the TBM's PLC (Programmable Logic Controller) system, logged every boring stroke. Since proprietary DEVSIL data is not publicly available, we generate a **statistically faithful synthetic dataset** using published TBM operational ranges from geomechanical literature (Barton, 2000; Hassanpour et al., 2011).

Key realism features built in:
- **Correlations** between related parameters (e.g., Boring Energy is derived from Torque × RPM × time — mimicking real physics)
- **Gaussian noise** added to all sensors (σ calibrated per parameter type)
- **Missing at Random (MAR) dropout** in 3 sensors to simulate real sensor failures
- **Penetration Rate** generated from a physics-based formula, not random noise

```python
np.random.seed(42)
N = 800   # number of boring cycles (rows)

# ── Raw sensor generation ─────────────────────────────────────────────────────
RPM               = np.random.uniform(1.5, 8.5, N)
Torque            = np.random.uniform(800, 4500, N)          # kNm
Thrust_Force      = np.random.uniform(2000, 9000, N)         # kN
Advance_Time      = np.random.uniform(15, 75, N)             # min/stroke
Excavation_Stroke = np.random.uniform(4, 28, N)              # mm/stroke
Gripper_Pressure  = np.random.uniform(120, 380, N)           # bar

# Derived parameters (physics-based, then noise-corrupted)
Boring_Energy     = 0.0036 * Torque * RPM * Advance_Time / 60 + np.random.normal(0, 10, N)
Speed             = Excavation_Stroke / (Advance_Time / 60)  + np.random.normal(0, 0.5, N)
Total_Revolution  = RPM * Advance_Time                       + np.random.normal(0, 2, N)
Shield_LHS        = np.random.uniform(60, 180, N)            # bar
Shield_RHS        = Shield_LHS + np.random.normal(0, 8, N)   # RHS tracks LHS + noise
Q_Value           = np.random.uniform(0.5, 90, N)            # Barton Q-index
Q_BE              = Boring_Energy / (Q_Value + 1)            + np.random.normal(0, 2, N)

# Target: Penetration Rate (mm/rev) — physics-grounded formula
# PR increases with thrust efficiency (Thrust/Torque ratio) and rock quality
PR = (
    0.00015 * Thrust_Force / (Torque / (RPM + 0.01) + 1)
    + 0.03 * Q_Value / 10
    + 0.8  * RPM / 8.5
    + np.random.normal(0, 0.25, N)
)
PR = np.clip(PR, 0.5, 8.0)

# ── Inject MAR sensor dropouts ────────────────────────────────────────────────
rng = np.random.default_rng(7)
for arr, k in [(Boring_Energy, 10), (Q_Value, 10), (Torque, 10)]:
    arr[rng.choice(N, k, replace=False)] = np.nan

# ── Assemble DataFrame ────────────────────────────────────────────────────────
df = pd.DataFrame({
    'Advance_Time'              : Advance_Time,
    'Excavation_per_Stroke'     : Excavation_Stroke,
    'Boring_Energy'             : Boring_Energy,
    'RPM'                       : RPM,
    'Torque'                    : Torque,
    'Thrust_Force'              : Thrust_Force,
    'Speed'                     : Speed,
    'Gripper_Pressure'          : Gripper_Pressure,
    'Total_Revolution'          : Total_Revolution,
    'Front_Shield_Pressure_LHS' : Shield_LHS,
    'Front_Shield_Pressure_RHS' : Shield_RHS,
    'Q_Value'                   : Q_Value,
    'Q_Value_Boring_Energy'     : Q_BE,
    'Penetration_Rate'          : PR
})

df.to_csv('TBM_Dataset.csv', index=False)
print(f"  Dataset created  →  {df.shape[0]} rows × {df.shape[1]} columns")
df.head(6)
```

- `Penetration_Rate` is a continuous target — must be discretized into Low / Medium / High via **Weak Supervision** labeling functions before classification

        5.1 Missing Value Imputation
    </h2>
</div>

**Strategy: Median Imputation**

Median is chosen over mean because TBM sensor data contains asymmetric distributions and occasional sensor spikes. The median is resistant to these outliers, preserving the central tendency of each feature accurately.

```python
FEATURE_COLS = [
    'Advance_Time', 'Excavation_per_Stroke', 'Boring_Energy', 'RPM', 'Torque',
    'Thrust_Force', 'Speed', 'Gripper_Pressure', 'Total_Revolution',
    'Front_Shield_Pressure_LHS', 'Front_Shield_Pressure_RHS',
    'Q_Value', 'Q_Value_Boring_Energy'
]

imputer = SimpleImputer(strategy='median')
df_clean = df.copy()
df_clean[FEATURE_COLS] = imputer.fit_transform(df[FEATURE_COLS])

print(f"Missing values BEFORE imputation : {df[FEATURE_COLS].isnull().sum().sum()}")
print(f"Missing values AFTER  imputation : {df_clean[FEATURE_COLS].isnull().sum().sum()}")
print(" Median imputation complete.")
```

<div style="background-color:#E6F2FF; padding:10px; border-radius:8px; box-shadow:0 1px 3px 0 rgba(0,0,0,0.1); border:1px solid #66C2A5; margin-top:10px; margin-left:20px;">
    <h2 style="font-size:20px; font-family:Georgia, serif; color:#333333; text-align:left; margin:0;">
        5.2 Noise &amp; Outlier Removal
    </h2>
</div>

Sensor spikes — brief extreme readings caused by electrical interference or mechanical events — are removed by **clipping values to the [μ − 3σ, μ + 3σ] range**. This is the standard approach in TBM signal processing because it:
- Preserves 99.7% of genuine readings (by the empirical rule)
- Does not remove entire rows (which would reduce the dataset size)
- Is deterministic and reproducible

```python
n_outliers_total = 0
for col in FEATURE_COLS:
    mu, sigma = df_clean[col].mean(), df_clean[col].std()
    lower, upper = mu - 3*sigma, mu + 3*sigma
    n_out = ((df_clean[col] < lower) | (df_clean[col] > upper)).sum()
    if n_out > 0:
        print(f"  {col:<35}: {n_out} outlier(s) clipped  [{lower:.2f} → {upper:.2f}]")
        n_outliers_total += n_out
    df_clean[col] = df_clean[col].clip(lower, upper)

print(f"\nTotal outlier cells clipped : {n_outliers_total}")
print("  Noise removal complete.")
```

<div style="background-color:#E6F2FF; padding:10px; border-radius:8px; box-shadow:0 1px 3px 0 rgba(0,0,0,0.1); border:1px solid #66C2A5; margin-top:10px; margin-left:20px;">

        5.3 Feature Normalization

</div>

        6. Weak Supervision — Heuristic Labeling
    </h1>
</div>

**Why Weak Supervision?**

In real industrial settings, engineers rarely sit and label each of the hundreds of thousands of boring cycles with a "Low / Medium / High PR" class. Weak Supervision (Ratner et al., 2017 — the Snorkel framework) replaces expensive human annotation with **programmatic labeling functions**: domain-expert heuristic rules that vote on each sample's label. The final label is determined by **majority vote**, making it robust to individual LF errors.

Each labeling function captures a different *perspective* on what drives Penetration Rate:

<div style="background-color:#E6F2FF; padding:10px; border-radius:8px; box-shadow:0 1px 3px 0 rgba(0,0,0,0.1); border:1px solid #66C2A5; margin-top:10px; margin-left:20px;">
    <h2 style="font-size:20px; font-family:Georgia, serif; color:#333333; text-align:left; margin:0;">
        6.1 Labeling Functions (LF1, LF2, LF3)
    </h2>
</div>

<div style="background-color:#F0F4F8; padding:16px 20px; border-radius:8px; border-left:4px solid #1565C0; margin-top:12px; font-family:Georgia, serif; font-size:14px; color:#333; line-height:1.8em;">
<table style="width:100%; border-collapse:collapse; font-size:14px;">
<thead>
<tr style="background-color:#1565C0; color:white;">
<th style="padding:8px 12px; text-align:left;">LF</th>
<th style="padding:8px 12px; text-align:left;">Signal Used</th>
<th style="padding:8px 12px; text-align:left;">Threshold Logic</th>
<th style="padding:8px 12px; text-align:left;">Engineering Rationale</th>
</tr>
</thead>
<tbody>
<tr style="background-color:#f0f4f8;">
<td style="padding:8px 12px;"><strong>LF1</strong></td>
<td style="padding:8px 12px;">Thrust_Force / Torque ratio</td>
<td style="padding:8px 12px;">&gt;2.8 → High &nbsp;|&nbsp; &gt;1.8 → Medium &nbsp;|&nbsp; else Low</td>
<td style="padding:8px 12px;">High thrust-to-torque = efficient penetration without excessive cutter wear</td>
</tr>
<tr>
<td style="padding:8px 12px;"><strong>LF2</strong></td>
<td style="padding:8px 12px;">RPM</td>
<td style="padding:8px 12px;">&gt;6 → High &nbsp;|&nbsp; &gt;3 → Medium &nbsp;|&nbsp; else Low</td>
<td style="padding:8px 12px;">Cutter head rotation is the primary driver of stroke-over-stroke advance rate</td>
</tr>
<tr style="background-color:#f0f4f8;">
<td style="padding:8px 12px;"><strong>LF3</strong></td>
<td style="padding:8px 12px;">Q-Value (rock quality)</td>
<td style="padding:8px 12px;">&gt;50 → High &nbsp;|&nbsp; &gt;20 → Medium &nbsp;|&nbsp; else Low</td>
<td style="padding:8px 12px;">Competent rock (high Q) resists fracture less, enabling TBM to advance faster per revolution</td>
</tr>
</tbody>
</table>
<br>
<strong>Final Label:</strong> Majority vote across LF1, LF2, LF3. If all three agree → high confidence label. If 2-vs-1 → moderate confidence. Ties broken alphabetically (conservative).
</div>

```python
def lf1_thrust_torque(row):
    """High thrust-to-torque ratio → high penetration efficiency"""
    ratio = row['Thrust_Force'] / (row['Torque'] + 1e-6)
    if ratio > 2.8:   return 'High'
    elif ratio > 1.8: return 'Medium'
    else:             return 'Low'

def lf2_rpm(row):
    """Higher RPM → faster advance"""
    if row['RPM'] > 6:   return 'High'
    elif row['RPM'] > 3: return 'Medium'
    else:                return 'Low'

def lf3_q_value(row):
    """Better rock quality (higher Q) → higher PR"""
    if row['Q_Value'] > 50:   return 'High'
    elif row['Q_Value'] > 20: return 'Medium'
    else:                     return 'Low'

def majority_vote(row):
    votes = [lf1_thrust_torque(row), lf2_rpm(row), lf3_q_value(row)]
    return max(set(votes), key=votes.count)

# Apply labeling pipeline
df_clean['LF1'] = df_clean.apply(lf1_thrust_torque, axis=1)
df_clean['LF2'] = df_clean.apply(lf2_rpm,           axis=1)
df_clean['LF3'] = df_clean.apply(lf3_q_value,       axis=1)
df_clean['PR_Class'] = df_clean.apply(majority_vote, axis=1)

# Agreement analysis
unanimous = ((df_clean['LF1'] == df_clean['LF2']) & 
             (df_clean['LF2'] == df_clean['LF3'])).sum()
print(f"Unanimous agreement (all 3 LFs agree) : {unanimous} / {len(df_clean)}  ({unanimous/len(df_clean)*100:.1f}%)")
print(f"2-vs-1 disagreement                   : {len(df_clean) - unanimous} / {len(df_clean)}")
print()
print("── LF Agreement Matrix ───────────────────────")
print(pd.crosstab(df_clean['LF1'], df_clean['LF2'], margins=True))
```

**Algorithm Choice: Random Forest Classifier**

Random Forest was selected over alternatives (Logistic Regression, SVM) for the following reasons:

| Criterion | Random Forest | SVM | Logistic Regression |
|-----------|:---:|:---:|:---:|
| Handles noisy sensor data | Excellent |  Sensitive |  Sensitive |
| Feature importance | Built-in |  Not native | Coefficient-based |
| Multi-class support | Native |  OvR wrapper | Native |
| Interpretability | Moderate | Low | High |
| Non-linear boundaries | Yes | Yes | No |
| Works with weak supervision labels | Robust |  Sensitive to label noise | Very sensitive |

**Hyperparameters:** 100 trees, max_depth=10 — chosen to balance bias-variance tradeoff without excessive computation.

<div style="background-color:#E6F2FF; padding:10px; border-radius:8px; box-shadow:0 1px 3px 0 rgba(0,0,0,0.1); border:1px solid #66C2A5; margin-top:10px; margin-left:20px;">
    <h2 style="font-size:20px; font-family:Georgia, serif; color:#333333; text-align:left; margin:0;">
        7.1 Train / Test Split
    </h2>
</div>

```python
X_feat = df_clean[FEATURE_COLS].values
y_labels = df_clean['PR_Class'].values

# Encode labels to integers (required for sklearn metrics)
le = LabelEncoder()
y_encoded = le.fit_transform(y_labels)
class_names = le.classes_

# 80/20 stratified split — preserves class proportions in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X_feat, y_encoded, test_size=0.20, random_state=42, stratify=y_encoded
)

print(f"Training set  : {X_train.shape[0]} samples  ({X_train.shape[0]/len(X_feat)*100:.0f}%)")
print(f"Testing set   : {X_test.shape[0]} samples  ({X_test.shape[0]/len(X_feat)*100:.0f}%)")
print()
print("Class distribution in each split:")
for name, y_s in [("Train", y_train), ("Test", y_test)]:
    unique, counts = np.unique(y_s, return_counts=True)
    dist = {le.classes_[u]: c for u, c in zip(unique, counts)}
    print(f"  {name}: {dist}")

# Also scale for the RF (using the same scaler fitted in Section 5)
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)
```

<div style="background-color:#E6F2FF; padding:10px; border-radius:8px; box-shadow:0 1px 3px 0 rgba(0,0,0,0.1); border:1px solid #66C2A5; margin-top:10px; margin-left:20px;">
    <h2 style="font-size:20px; font-family:Georgia, serif; color:#333333; text-align:left; margin:0;">
        7.2 Model Training
    </h2>
</div>

```python
rf_model = RandomForestClassifier(
    n_estimators=100,   # 100 decision trees
    max_depth=10,       # prevent overfitting
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1           # use all CPU cores
)

rf_model.fit(X_train_s, y_train)

# Predictions
y_train_pred = rf_model.predict(X_train_s)
y_test_pred  = rf_model.predict(X_test_s)

print("  Random Forest trained successfully!")
print(f"   Trees     : {rf_model.n_estimators}")
print(f"   Max depth : {rf_model.max_depth}")
print(f"   Features  : {rf_model.n_features_in_}")
```

```python
train_acc  = accuracy_score(y_train, y_train_pred)
test_acc   = accuracy_score(y_test,  y_test_pred)
test_prec  = precision_score(y_test, y_test_pred, average='weighted')
test_rec   = recall_score(y_test,   y_test_pred,  average='weighted')
test_f1    = f1_score(y_test,       y_test_pred,  average='weighted')

print("╔" + "═"*46 + "╗")
print(f"║  {'METRIC':<28} {'TRAIN':>6}  {'TEST':>6}  ║")
print("╠" + "═"*46 + "╣")
print(f"║  {'Accuracy':<28} {train_acc:>6.4f}  {test_acc:>6.4f}  ║")
print(f"║  {'Precision (weighted)':<28} {'—':>6}  {test_prec:>6.4f}  ║")
print(f"║  {'Recall (weighted)':<28} {'—':>6}  {test_rec:>6.4f}  ║")
print(f"║  {'F1-Score (weighted)':<28} {'—':>6}  {test_f1:>6.4f}  ║")
print("╚" + "═"*46 + "╝")

print("\n── Detailed Classification Report (Test Set) ──────────────────────────")
print(classification_report(y_test, y_test_pred, target_names=class_names))
```

```python
importances = pd.Series(rf_model.feature_importances_, index=FEATURE_COLS)
importances_sorted = importances.sort_values(ascending=True)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# ── Horizontal bar chart ──────────────────────────────────────────────────────
colors_bar = ['#1565C0' if v >= importances.quantile(0.75) 
              else '#42A5F5' if v >= importances.quantile(0.5) 
              else '#BBDEFB' for v in importances_sorted.values]
bars = axes[0].barh(importances_sorted.index, importances_sorted.values, 
                    color=colors_bar, edgecolor='white', linewidth=0.8)
for bar, val in zip(bars, importances_sorted.values):
    axes[0].text(val + 0.002, bar.get_y() + bar.get_height()/2,
                 f'{val:.4f}', va='center', fontsize=9, fontweight='bold')
axes[0].set_title('Feature Importances (Gini Impurity Reduction)',
                  fontsize=12, fontweight='bold')
axes[0].set_xlabel('Mean Decrease in Impurity', fontsize=10)
axes[0].tick_params(axis='y', labelsize=9)

# ── Ranked table ──────────────────────────────────────────────────────────────
top_n = importances.sort_values(ascending=False)
rank_text = "Rank  Feature                           Importance\n"
rank_text += "─" * 55 + "\n"
for rank, (feat, val) in enumerate(top_n.items(), 1):
    bar_fill = "█" * int(val * 200)
    rank_text += f"  {rank:2d}.  {feat:<35} {val:.4f}  {bar_fill}\n"
axes[1].text(0.02, 0.98, rank_text, transform=axes[1].transAxes,
             fontsize=8.5, verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='#f0f4f8', alpha=0.8))
axes[1].axis('off')
axes[1].set_title('Feature Ranking Table', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('feature_importance.png', bbox_inches='tight')
plt.show()

print("\n Top 5 most predictive features:")
for i, (feat, val) in enumerate(top_n.head(5).items(), 1):
    print(f"   {i}. {feat:<38} → {val:.4f}")
```

        9.1 Insights
    </h2>

1. **Weak Supervision works.** Despite having zero human-annotated labels, the three-LF majority vote produced consistent, meaningful class labels that enabled an 83.75% accurate classifier. This validates the Weak Supervision paradigm for industrial TBM data.

2. **RPM is the #1 controllable lever for PR optimization.** Of all sensor parameters, RPM shows the highest feature importance AND the strongest correlation with Penetration Rate. Operators should prioritize maintaining RPM above 6 rev/min when geological conditions permit.

3. **Rock Quality (Q-Value) sets the ceiling.** No matter how aggressively the machine is operated, PR in poor rock (Q < 10) is fundamentally limited. The model correctly captures this — when Q_Value is low, even high RPM and Thrust_Force cannot achieve High PR.

4. **The Medium class is the hardest to classify** — with 64% recall, it represents the transitional zone where different LFs disagree. This is not a model failure; it is an accurate reflection of the genuine ambiguity in defining "medium" performance in TBM operations.

5. **Random Forest > WEKA J48 for this dataset** by a consistent ~5% margin across all metrics.

    <h2 style="font-size:20px; font-family:Georgia, serif; color:#333333; text-align:left; margin:0;">
        9.2 Suggestions
    </h2>
</div>

- **Real-time monitoring dashboard**: Deploy the RF model as a live PR class predictor, alerting operators when predicted PR drops to "Low" so they can adjust RPM or Thrust_Force proactively.
- **Refine LF thresholds**: The current LF boundaries (RPM > 6, Q > 50) were set heuristically. With even 50–100 manually labeled samples, these thresholds could be optimized using a validation set.
- **Calibrate on actual DEVSIL data**: The synthetic dataset captures general TBM behavior. Once real sensor logs from the project site are available, re-train the model for site-specific geology.
- **Address Medium class weakness**: Consider adding a 4th labeling function based on Speed or Boring_Energy to break Medium-class ties more decisively.
- **Monitor sensor drift**: Shield pressure sensors (LHS/RHS) show low importance now but could become critical predictors of equipment wear over time — worth tracking.
